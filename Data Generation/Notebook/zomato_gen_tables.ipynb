## Build Menu Dataframe

```python
def classify_veg_type(item):

    item = item.lower()

    if any(x in item for x in ["chicken", "mutton", "fish", "prawn"]):
        return "nonveg"

    elif "egg" in item:
        return "egg"

    else:
        return "veg"
```

```python
def extract_cuisine_and_course(category, cuisines_list):
    
    category_lower = category.lower()
    
    cuisine_match = None
    
    for cuisine in cuisines_list:
        if category_lower.startswith(cuisine.lower()):
            cuisine_match = cuisine
            break
    
    if cuisine_match is None:
        raise ValueError(f"Unknown cuisine in category: {category}")
    
    course = category_lower.replace(cuisine_match.lower(), "").strip()
    
    return cuisine_match, course
```

```python
menu_rows = []
item_id = 0

for category, items in cuisine_items.items():

    cuisine, course = extract_cuisine_and_course(category, cuisines)

    for item in items:

        veg_type = classify_veg_type(item)

        menu_rows.append({
            "item_id": item_id,
            "item_name": item,
            "cuisine": cuisine,
            "course": course,
            "veg_type": veg_type
        })

        item_id += 1

menu_df = pd.DataFrame(menu_rows)
```

## Generate Orders Table

#### Initialzing the Parameters

```python
P_MAIN = 0.85

P_APPETIZER_IF_MAIN = 0.25
P_COMPLEMENT_IF_MAIN = 0.80
P_EXTRA_IF_MAIN = 0.60
P_DESSERT_IF_MAIN = 0.25

P_APPETIZER_IF_NO_MAIN = 0.65
P_COMPLEMENT_IF_NO_MAIN = 0.10
P_EXTRA_IF_NO_MAIN = 0
P_DESSERT_IF_NO_MAIN = 0.10
```

```python
# Item selection Logic

prob_table_if_main = {
    "mains": 0.05,  # allow for multiple mains in some cases
    "appetizers": P_APPETIZER_IF_MAIN,
    "complements": P_COMPLEMENT_IF_MAIN,
    "extras": P_EXTRA_IF_MAIN,
    "desserts & beverages": P_DESSERT_IF_MAIN
}

prob_table_if_no_main = {
    "mains": P_MAIN,
    "appetizers": P_APPETIZER_IF_NO_MAIN,
    "complements": P_COMPLEMENT_IF_NO_MAIN,
    "extras": P_EXTRA_IF_NO_MAIN,
    "desserts & beverages": P_DESSERT_IF_NO_MAIN
}
```

```python
usage_weights = {"high": 0.6,
                 "medium": 0.3,
                 "low": 0.2}
```

### Defining function for orders and recommendations

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
def ordered_items(cuisine, items_number):

    cuisine_menu = menu_df[menu_df["cuisine"] == cuisine]

    selected_items = []
    selected_item_ids = set()

    main_selected = False

    while len(selected_items) < items_number:
        # choose correct probability table
        if main_selected:
            prob_table = prob_table_if_main
        else:
            prob_table = prob_table_if_no_main

        categories = list(prob_table.keys())
        probs = np.array(list(prob_table.values()))

        # normalize
        probs = probs / probs.sum()

        chosen_course = np.random.choice(categories, p = probs)
        # sample item from category
        recommendation_choice = cuisine_menu[
            (cuisine_menu["course"] == chosen_course) &
            (~cuisine_menu["item_id"].isin(selected_item_ids))
        ]

        if len(recommendation_choice) == 0:
            continue

        item = recommendation_choice.sample(1)["item_id"].iloc[0]

        selected_items.append(item)
        selected_item_ids.add(item)

        if chosen_course == "main":
            main_selected = True

    return selected_items
```

```python
def recommended_items(cuisine, items_ordered, n_recommend = 3):

    cuisine_menu = menu_df[menu_df["cuisine"] == cuisine]

    # Removing already ordered items
    candidate_menu = cuisine_menu[~cuisine_menu["item_id"].isin(items_ordered)]

    # Identify what courses are already ordered
    ordered_courses = cuisine_menu[cuisine_menu["item_id"].isin(items_ordered)]["course"].unique().tolist()
    has_main = "main" in ordered_courses

    if has_main:
        rec_probs = prob_table_if_main
    else:
        rec_probs = prob_table_if_no_main

    recommended_ids = []
    recommended_names = []


    while len(recommended_ids) < n_recommend:
        categories = list(rec_probs.keys())
        probs = np.array(list(rec_probs.values()))
        probs = probs / probs.sum()

        chosen_category = np.random.choice(categories, p=probs)

        subset = candidate_menu[
            (candidate_menu["course"] == chosen_category) &
            (~candidate_menu["item_id"].isin(recommended_ids))
        ]

        if len(subset) == 0:
            continue

        item_id = subset.sample(1)["item_id"].iloc[0]
        item_name = subset[subset["item_id"] == item_id]["item_name"].iloc[0]

        recommended_ids.append(item_id)
        recommended_names.append(item_name)

    return recommended_ids, recommended_names
```

```python
item_course_map = dict(zip(menu_df["item_id"], menu_df["course"]))
item_cuisine_map = dict(zip(menu_df["item_id"], menu_df["cuisine"]))

# Optional but recommended
item_popularity = {item: np.random.uniform(0.2, 1.0)
                   for item in menu_df["item_id"]}
```

```python
complement_rules = {
    "main": ["complement", "extras", "dessert"],
    "appetizer": ["main", "dessert"],
    "complement": ["main", "extras"],
    "extras": ["main", "complement"],
    "dessert": ["main", "appetizer"]
}
```

```python
def recommendation_click_probability(
    cust,
    ordered_ids,
    rec_id,
    position
):

    ordered_courses = {
        item_course_map[i] for i in ordered_ids
    }

    rec_course = item_course_map[rec_id]
    rec_cuisine = item_cuisine_map[rec_id]

    # Complement score
    complement_score = max(
        [rec_course in complement_rules.get(course, [])
        for course in ordered_courses],
        default = False
    )

    complement_score = 0.6 if complement_score else 0.15

    # Cuisine match score
    cuisine_match = 0.6 if rec_cuisine in cuisines else 0.3

    # User engagement
    usage_score = {
        "low": 0.2,
        "medium": 0.4,
        "high": 0.7
    }[cust.app_usage_frequency]

    # Position bias
    position_score = 1/ (position + 1)

    # Popularity
    popularity_score = item_popularity.get(rec_id, 0.5) * 0.6

    # Basket size influence
    basket_score = min(len(ordered_ids)/ 8, 0.6)

    intercept = -2.2

    # Final score
    score = (
        intercept +
        1.0 * complement_score +
        0.7 * cuisine_match +
        0.9 * usage_score +
        0.8 * position_score +
        0.6 * popularity_score +
        0.5 * basket_score +
        np.random.normal(0, 0.4)
    )

    return sigmoid(score)
```

```python
def simulate_recommendation_click(cust, ordered_ids, recommendation_ids):

    clicked = []

    for pos, rec_id in enumerate(recommendation_ids, start = 1):

        prob = recommendation_click_probability(cust, ordered_ids, rec_id, pos)

        if np.random.rand() < prob:
            if np.random.rand() < 0.55:
                clicked.append(1)
            else:
                clicked.append(0)
        else:
            clicked.append(0)

    return clicked
```

### Using the defined functions to simulate a logical $\text{Orders Table}$

```python
orders = []

for i in range(N_ORDERS):

    # A random customer is chosen to place an order, with higher probability for more frequent app users
    cust = customers_df.sample(1,
                               weights = customers_df["app_usage_frequency"].map(usage_weights)
                               ).iloc[0]
    
    order_time = fake.date_time_between(start_date = "-1y", end_date = "now").strftime("%H:%M:%S")

    # Extract cuisine probability vector
    probs = np.array(json.loads(cust["cuisine_preference"]))
    cuisine = np.random.choice(cuisines, p = probs)

    # Choose cuisine using those probabilities
    cuisine = np.random.choice(cuisines, p = probs)
    items_number = int(min(max(1, np.random.normal(cust["avg_items"], 3)), 12))
    items_ordered_ids = ordered_items(cuisine, items_number)

    recommendation_ids, recommended_names = recommended_items(cuisine, items_ordered_ids, n_recommend = 3)

    recommendation_added = False

    if recommendation_ids:
        match = int(cuisine == cust.cuisine_preference.split(',')[0].strip())
        score = (
            1.5 * match +
            0.5 * (cust.app_usage_frequency == "high") +
            np.random.normal(0,1)
        )


    recommendation_added = simulate_recommendation_click(cust, items_ordered_ids, recommendation_ids)
    
    orders.append({
        "order_id": 8000000 + i,
        "customer_id": cust.customer_id,
        "order_time": order_time,
        "cuisine": cuisine,
        "items_ordered_id": items_ordered_ids,
        "items_ordered_name": menu_df[menu_df["item_id"].isin(items_ordered_ids)]["item_name"].tolist(),
        "recommendation_shown_id": recommendation_ids,
        "recommendation_shown_name": recommended_names,
        "recommendation_added_1": recommendation_added[0],
        "recommendation_added_2": recommendation_added[1],
        "recommendation_added_3": recommendation_added[2],
    })

orders_df = pd.DataFrame(orders)
```
