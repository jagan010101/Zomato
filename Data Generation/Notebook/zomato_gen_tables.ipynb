### Building scripts for labelling the items in the menu

```python
def classify_veg_type(item):
    item = item.lower()
    
    if any(x in item for x in ["chicken", "mutton", "fish", "prawn"]):
        return "nonveg"

    elif "egg" in item:
        return "egg"

    else:
        return "veg"
```

```python
def extract_cuisine_and_course(category, cuisines_list):
    category_lower = category.lower()
    cuisine_match = None
    
    for cuisine in cuisines_list:
        if category_lower.startswith(cuisine.lower()):
            cuisine_match = cuisine
            break
    
    if cuisine_match is None:
        raise ValueError(f"Unknown cuisine in category: {category}")
    
    course = category_lower.replace(cuisine_match.lower(), "").strip()
    
    return cuisine_match, course
```

### Generating prices for the items in the menu

```python
course_base_price = {
    "mains": 150,
    "appetizers": 250,
    "complements": 100,
    "extras": 50,
    "desserts & beverages": 110
}

cuisine_multiplier = {
    "North Indian": 1.0,
    "South Indian": 0.85,
    "Biryani": 1.2,
    "Chinese": 1.1,
    "Italian": 1.4,
    "Fast Food": 0.9,
    "Healthy": 1.2
}

cuisine_course_override = {
    ("Italian", "mains"): 250,
    ("Biryani", "mains"): 200,
    ("North Indian", "complements"): 200,
    ("Chinese", "complements"): 200,
    ("Italian", "complements"): 80,
    ("South Indian", "complements"): 80,
    ("Italian", "appetizers"): 150,
    ("South Indian", "appetizers"): 150,
}

tier_multiplier = {
    "Budget": 0.85,
    "Premium": 1.25
}
```

```python
def item_multiplier(item_name):
    name = item_name.lower()
    modifier = 1.0

    if "plain" in name:
        modifier *= 0.90

    if "butter" in name:
        modifier *= 1.15

    if "cheese" in name:
        modifier *= 1.20

    if "paneer" in name:
        modifier *= 1.18

    if "chicken" in name:
        modifier *= 1.25

    if "mutton" in name:
        modifier *= 1.45

    if "fish" in name:
        modifier *= 1.60

    if "egg" in name:
        modifier *= 1.05

    return modifier
```

```python
def get_base_price(cuisine, course):
    # Check override first
    if (cuisine, course) in cuisine_course_override:
        return cuisine_course_override[(cuisine, course)]

    base = course_base_price[course]
    mult = cuisine_multiplier.get(cuisine, 1.0)

    return base * mult


def get_multiplier(item_name):
    item_modifier = item_multiplier(item_name)

    return item_modifier
```

```python
def generate_popularity(menu, seed = 42):
    np.random.seed(seed)
    menu["Popularity Score"] = np.random.lognormal(mean = 0.0, sigma = 1.0, size = len(menu))
    
    # Normalize to 0–1
    menu["Popularity Score"] = (menu["Popularity Score"] - 
                                menu["Popularity Score"].min()) / (menu["Popularity Score"].max() - 
                                                                   menu["Popularity Score"].min())
    
    return menu
```

### Generating customer's preferences to help build the $\text{Orders Table}$

```python
def generate_veg_pref():
    r = np.random.rand()
    
    # 20% pure veg
    if r < 0.20:
        return [1.0, 0.0, 0.0]

    # 15% veg + egg only
    elif r < 0.35:
        veg_egg = np.random.dirichlet([3, 1])  # bias toward veg
        return [veg_egg[0], 0.0, veg_egg[1]]

    # remaining 65% mixed
    else:
        mixed = np.random.dirichlet([2, 3, 1])  # slight nonveg bias
        return mixed.tolist()
```

```python
def generate_time_pref():
    r = np.random.rand()

    # 12% Late-night users
    if r < 0.12:
        probs = np.random.dirichlet([4, 2, 1, 3, 1])  # late night dominant

    # 88% Non-late-night users
    else:
        probs = np.random.dirichlet([0.2, 3, 2, 4, 2])
        probs[0] = 0
        probs = probs / probs.sum()  # re-normalize to sum to 1

    return probs.tolist()
```

```python
city_cuisine_alpha = {
    "Delhi": [4, 1, 2, 1, 1, 1, 1],
    "Mumbai": [3, 3, 2, 4, 1, 1, 4],
    "Bangalore": [2, 4, 2, 2, 1, 2, 4],
    "Chennai": [1, 4, 2, 1, 2, 1, 2],
}


def generate_cuisine_pref(city, age):
    # Base city alpha
    base_alpha = np.array(city_cuisine_alpha.get(city, np.ones(len(cuisines))), dtype = float)

    alpha = base_alpha.copy()

    # AGE EFFECTS

    if age < 25:    # boost fast food, biryani, italian
        alpha[cuisines.index("Fast Food")] += 2
        alpha[cuisines.index("Biryani")] += 2
        alpha[cuisines.index("Italian")] += 1

    elif 25 <= age < 40:    # mild boost to fast food, chinese, biryani and italian
        alpha[cuisines.index("Fast Food")] += 1.25
        alpha[cuisines.index("Chinese")] += 1
        alpha[cuisines.index("Biryani")] += 1.25
        alpha[cuisines.index("Italian")] += 1

    elif 40 <= age < 55:    # mild reduction in fast food
        alpha[cuisines.index("North Indian")] += 2
        alpha[cuisines.index("South Indian")] += 2
        alpha[cuisines.index("Fast Food")] *= 0.75

    else:   # significant reduction fast food
        alpha[cuisines.index("South Indian")] += 2
        alpha[cuisines.index("Fast Food")] *= 0.15

    # Ensure no zeros
    alpha = np.maximum(alpha, 0.1)

    probs = np.random.dirichlet(alpha)

    return probs.tolist()
```

```python
base_course_alpha = {
    "mains": 3,
    "appetizers": 2,
    "complements": 2,
    "extras": 1,
    "desserts & beverages": 2
}


def generate_course_pref(customer):
    alpha = base_course_alpha.copy()
    age = customer["Age"]
    city = customer["City"]
    avg_value = customer["Avg Value per Item"]
    time_probs = np.array(json.loads(customer["Time Preference"]))
    dominant_time = time_categories[np.argmax(time_probs)]

    # Time effect
    if dominant_time == "dinner":
        alpha["mains"] += 2
        alpha["complements"] += 1
    if dominant_time == "snack":
        alpha["desserts & beverages"] += 2
        alpha["appetizers"] += 1
    if dominant_time == "late night":
        alpha["desserts & beverages"] += 2

    # Age effect
    if age < 25:
        alpha["extras"] += 1
        alpha["desserts & beverages"] += 1
    if age > 40:
        alpha["mains"] += 1

    # High spending effect
    if avg_value > 400:
        alpha["appetizers"] += 1
        alpha["desserts & beverages"] += 1

    # City effect
    if city == "Mumbai":
        alpha["appetizers"] += 1
    if city == "Chennai":
        alpha["mains"] += 1

    alpha_vector = [alpha[c] for c in courses]

    return np.random.dirichlet(alpha_vector).tolist()
```

```python
def item_score(row, cust, order_time_category):
    score = 1.0

    # Veg Compatibility
    veg_probs = np.array(json.loads(cust["Veg Nonveg Egg Preference"]))
    veg_type = row["Veg Type"]

    if veg_type == "veg":
        score *= (0.1 + veg_probs[0])
    elif veg_type == "nonveg":
        score *= (0.1 + veg_probs[1])
    else:
        score *= (0.1 + veg_probs[2])

    # Prevent zero collapse
    score = max(score, 1e-6)


    # Price Compatibility (smooth Gaussian)
    avg_value = cust["Avg Value per Item"]
    price = row["Base Price"]

    sigma = avg_value * 0.4  # adaptive spread
    price_weight = np.exp(-((price - avg_value) ** 2) / (2 * sigma ** 2))
    score *= price_weight


    # Age Effects
    age = cust["Age"]

    if age < 25:
        if row["Course"] in ["desserts & beverages", "appetizers"]:
            score *= 1.5
        if row["Cuisine"] == "Fast Food":
            score *= 1.3
        if row["Cuisine"] == "Biryani":
            score *= 1.5
        if row["Cuisine"] == "Healthy Food":
            score *= 1.2


    elif age > 45:
        if row["Cuisine"] == "South Indian":
            score *= 1.3
        if row["Cuisine"] == "North Indian":
            score *= 1.3
        if row["Course"] == "desserts & beverages":
            score *= 0.45
        if row["Course"] == "appetizers":
            score *= 0.65

    
    # Time-of-Day Effects
    if order_time_category == "breakfast":
        if row["Course"] == "mains":
            score *= 1.3

    if order_time_category == "snack":
        if row["Course"] in ["desserts & beverages", "appetizers"]:
            score *= 1.4
    
    if order_time_category == "lunch":
        if row["Course"] == "mains":
            score *= 1.5
        if row["Course"] == "complements":
            score *= 1.2

    if order_time_category == "dinner":
        if row["Course"] == "mains":
            score *= 1.5
        if row["Course"] == "complements":
            score *= 1.2

    if order_time_category == "late night":
        if row["Course"] == "desserts & beverages":
            score *= 1.6


    # City Preference
    cuisine_probs = np.array(json.loads(cust["Cuisine Preference"]))
    cuisine_index = cuisines.index(row["Cuisine"])
    score *= (0.5 + cuisine_probs[cuisine_index])

    
    # Course Preference Reinforcement
    course_probs = np.array(json.loads(cust["Course Preference"]))
    course_index = courses.index(row["Course"])
    score *= (0.5 + course_probs[course_index])

    
    # Small Random Noise (heterogeneity)
    score *= np.random.normal(1.0, 0.05)

    return max(score, 1e-8)
```

### Rounding to 2 decimal places

```python
def round_and_fix_probs(prob_list, decimals=2):

    probs = np.array(prob_list)
    rounded = np.round(probs, decimals)

    diff = 1.0 - rounded.sum()

    idx = np.argmax(rounded)
    rounded[idx] += diff

    rounded = np.round(rounded, decimals)

    return [float(f"{x:.2f}") for x in rounded]
```

### Defining the necessary functions for orders generation

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
def generate_time_from_pref(cust):
    time_probs = np.array(json.loads(cust["Time Preference"]))
    category = np.random.choice(time_categories, p = time_probs)

    start_hour, end_hour = time_ranges[category]

    # handle midnight wrap
    if start_hour > end_hour:
        if random.random() < 0.5:
            hour = random.randint(start_hour, 23)
        else:
            hour = random.randint(0, end_hour)
    else:
        # normal peak-based generation
        peak = (start_hour + end_hour) // 2
        hour = int(np.clip(np.random.normal(peak, 1), start_hour, end_hour))

    minute = random.randint(0, 59)
    second = random.randint(0, 59)

    return hour, minute, second
```

```python
def ordered_items(cuisine, items_number, customer):
    selected_items = []
    selected_item_ids = set()

    veg_probs = np.array(json.loads(customer["Veg Nonveg Egg Preference"]))
    course_probs = np.array(json.loads(customer["Course Preference"]))

    for _ in range(items_number):

        # Sample course using customer preference
        course = np.random.choice(courses, p = course_probs)

        # Filter menu by cuisine + course
        subset = menu_df[
            (menu_df["Cuisine"] == cuisine) &
            (menu_df["Course"] == course) &
            (~menu_df["Item ID"].isin(selected_item_ids))
        ]

        if subset.empty:
            continue

        # Score items
        scores = subset.apply(lambda row: item_score(row, customer, time_categories), axis = 1).values

        if scores.sum() <= 1e-12:
            probs = np.ones(len(subset)) / len(subset)
        
        else:
            probs = scores / scores.sum()

        chosen_item = np.random.choice(subset["Item ID"], p = probs)

        selected_items.append(chosen_item)
        selected_item_ids.add(chosen_item)

    return selected_items
```

```python
def recommended_items(cuisine, items_ordered, n_recommend = 3):
    cuisine_menu = menu_df[menu_df["Cuisine"] == cuisine]

    # Removing already ordered items
    candidate_menu = cuisine_menu[~cuisine_menu["Item ID"].isin(items_ordered)]

    # Identify what courses are already ordered
    ordered_courses = cuisine_menu[cuisine_menu["Item ID"].isin(items_ordered)]["Course"].unique().tolist()
    has_main = "mains" in ordered_courses

    if has_main:
        rec_probs = prob_table_if_main
    else:
        rec_probs = prob_table_if_no_main

    recommended_ids = []
    recommended_names = []


    while len(recommended_ids) < n_recommend:
        categories = list(rec_probs.keys())
        probs = np.array(list(rec_probs.values()))
        probs = probs / probs.sum()

        chosen_category = np.random.choice(categories, p = probs)

        subset = candidate_menu[
            (candidate_menu["Course"] == chosen_category) &
            (~candidate_menu["Item ID"].isin(recommended_ids))
        ]

        if len(subset) == 0:
            continue

        item_id = subset.sample(1)["Item ID"].iloc[0]
        item_name = subset[subset["Item ID"] == item_id]["Item Name"].iloc[0]

        recommended_ids.append(item_id)
        recommended_names.append(item_name)

    return recommended_ids, recommended_names
```

```python
item_course_map = dict(zip(menu_df["Item ID"], menu_df["Course"]))
item_cuisine_map = dict(zip(menu_df["Item ID"], menu_df["Cuisine"]))

item_popularity = {item: np.random.uniform(0.2, 1.0) for item in menu_df["Item ID"]}
```

```python
complement_rules = {
    "mains": ["appetizers", "complements", "extras", "desserts & beverages"],
    "appetizers": ["mains", "desserts & beverages"],
    "complements": ["mains"],
    "extras": ["mains", "appetizers"],
    "desserts & beverages": ["mains", "appetizers"]
}
```

```python
def recommendation_click_probability(cust, ordered_ids, rec_id, position):
    ordered_courses = {item_course_map[i] for i in ordered_ids}

    rec_course = item_course_map[rec_id]
    rec_cuisine = item_cuisine_map[rec_id]

    # Complement score
    complement_score = max(
        [rec_course in complement_rules.get(course, [])
        for course in ordered_courses],
        default = False
    )

    complement_score = 0.6 if complement_score else 0.15

    # Cuisine match score
    cuisine_match = 0.6 if rec_cuisine in cuisines else 0.3

    # User engagement
    usage_score = {
        "low": 0.2,
        "medium": 0.4,
        "high": 0.7
    }[cust["App Usage Frequency"]]

    # Position bias
    position_score = 1/ (position + 1)

    # Popularity
    popularity_score = item_popularity.get(rec_id, 0.5) * 0.6

    # Basket size influence
    basket_score = min(len(ordered_ids)/ 8, 0.6)

    intercept = -2.2

    # Final score
    score = (
        intercept +
        1.0 * complement_score +
        0.7 * cuisine_match +
        0.9 * usage_score +
        0.8 * position_score +
        0.6 * popularity_score +
        0.5 * basket_score +
        np.random.normal(0, 0.4)
    )

    return sigmoid(score)
```

```python
def recommendation_click_probability(cust, ordered_ids, rec_id, position, menu, complement_rules):
    ordered_df = menu[menu["Item ID"].isin(ordered_ids)]
    rec_row = menu[menu["Item ID"] == rec_id].iloc[0]

    ordered_courses = set(ordered_df["Course"])
    ordered_cuisines = set(ordered_df["Cuisine"])

    rec_course = rec_row["Course"]
    rec_cuisine = rec_row["Cuisine"]

    popularity_score = rec_row["Popularity Score"]

    basket_size = len(ordered_ids)

    # Complement score
    num_complements = sum(
        rec_course in complement_rules.get(course, [])
        for course in ordered_courses
    )

    complement_score = num_complements / max(len(ordered_courses), 1)

    # Cuisine score
    cuisine_score = 1.0 if rec_cuisine in ordered_cuisines else 0.0

    # Usage score
    usage_score_map = {
        "low": 0.3,
        "medium": 0.6,
        "high": 1.0
    }
    usage_score = usage_score_map.get(
        cust.get("App Usage Frequency", "medium"),
        0.6
    )

    # Position bias
    position_score = 1 / np.log2(position + 2)

    # Basket stage
    basket_score = min(basket_size / 8.0, 1.0)

    # Logistic model
    intercept = -2.5

    score = (
        intercept
        + 1.2 * complement_score
        + 0.9 * cuisine_score
        + 0.8 * usage_score
        + 1.0 * position_score
        + 0.7 * popularity_score
        + 0.6 * basket_score
    )

    return 1 / (1 + np.exp(-score))
```

```python
def simulate_recommendation_click(cust, ordered_ids, recommendation_ids):
    clicked = []

    # Check if order is strictly veg
    ordered_veg_types = [item_veg_type_map[i] for i in ordered_ids]

    all_veg_order = all(v == "veg" for v in ordered_veg_types)

    for pos, rec_id in enumerate(recommendation_ids, start = 1):

        rec_veg_type = item_veg_type_map[rec_id]

        # On veg days, customer won't click on non veg recommendation
        if all_veg_order and (rec_veg_type == "nonveg" or rec_veg_type == "egg"):
            clicked.append(0)
            continue

        prob = recommendation_click_probability(cust, ordered_ids, rec_id, pos, menu_df, complement_rules)

        if np.random.rand() < prob:
            if np.random.rand() < 0.55:
                clicked.append(1)
            else:
                clicked.append(0)
        else:
            clicked.append(0)

    return clicked
```

### Train-test split based on chronology

```python
def chronological_split(df, time_col = "Order Time", test_ratio = 0.2):
    df = df.sort_values(time_col)

    cutoff_index = int(len(df) * (1 - test_ratio))
    cutoff_time = df.iloc[cutoff_index][time_col]

    train = df[df[time_col] <= cutoff_time].copy()
    test  = df[df[time_col] > cutoff_time].copy()

    return train, test
```

### Calculating order metrics

```python
def add_order_metrics(orders_df, menu_df):
    df = orders_df.copy()

    # Ensure items_ordered is a list
    def ensure_list(x):
        if isinstance(x, list):
            return x
        return ast.literal_eval(x)

    df["Items Ordered ID"] = df["Items Ordered ID"].apply(ensure_list)

    # Explode
    exploded = df.explode("Items Ordered ID")

    # Merge prices
    exploded = exploded.merge(
        menu_df[["Item ID", "Budget Price"]],
        left_on = "Items Ordered ID",
        right_on = "Item ID",
        how = "left"
    )

    exploded["price"] = exploded["Budget Price"].fillna(0)

    # Aggregate
    summary = (
        exploded.groupby("Order ID")
        .agg(
            order_value = ("Budget Price", "sum"),
            num_items = ("Items Ordered ID", "count")
        )
        .reset_index()
    )

    # Merge back
    df = df.merge(summary, on = "Order ID", how = "left")

    # Rename the columns
    df = df.rename(columns = {
        "order_value": "Order Value",
        "num_items": "Number of Items"
    })
    
    return df
```

### Calculating popularity score for items in menu

```python
def compute_item_popularity(train_df):
    df = train_df.copy()

    # Ensure list format
    def ensure_list(x):
        if isinstance(x, list):
            return x
        return ast.literal_eval(x)

    df["Items Ordered ID"] = df["Items Ordered ID"].apply(ensure_list)

    # Explode items
    exploded = df.explode("Items Ordered ID")

    # Count frequency
    item_counts = (
        exploded
        .groupby("Items Ordered ID")
        .size()
        .reset_index(name = "Order Count")
    )

    return item_counts
```

```python
def normalize_popularity(item_counts):
    item_counts["Popularity Score"] = (
        item_counts["Order Count"] - 
        item_counts["Order Count"].min()) / (item_counts["Order Count"].max() - 
                                             item_counts["Order Count"].min())

    return item_counts
```

```python
item_counts = compute_item_popularity(orders_df)
item_counts = normalize_popularity(item_counts)

menu_df = menu_df.merge(
    item_counts,
    left_on = "Item ID",
    right_on = "Items Ordered ID",
    how = "left"
)

menu_df["Order Count"] = menu_df["Order Count"].fillna(0)
menu_df["Popularity Score"] = menu_df["Popularity Score"].fillna(0)
```
